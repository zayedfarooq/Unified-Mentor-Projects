```python
#Checking out the results at the world cup
def Winner(df1_without_duplicates):
    if df1_without_duplicates['Home Team Goals'] > df1_without_duplicates['Away Team Goals']:
        return 'Home team won'
    elif df1_without_duplicates['Home Team Goals'] < df1_without_duplicates['Away Team Goals']:
        return 'Away Team won'
    elif df1_without_duplicates['Home Team Goals'] == df1_without_duplicates['Away Team Goals']:
        return 'Draw'
```

```python
#Creating a new column for the results
df1_without_duplicates['Winner']=df1_without_duplicates.apply(lambda x:Winner(x),axis=1)
df1_without_duplicates.head()
```
