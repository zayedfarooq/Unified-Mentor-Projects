```python
# Group by 'Order Priority' and calculate the mean for 'Total Revenue', 'Total Cost', and 'Total Profit'
grouped_data = df.groupby('Year')[['Total Revenue', 'Total Cost', 'Total Profit']].mean()

# Plotting the side-by-side bar graph
bar_width = 0.2
index = np.arange(len(grouped_data))

# Create the bars for Total Revenue, Total Cost, and Total Profit
plt.figure(figsize=(14, 6))
plt.bar(index, grouped_data['Total Revenue'], bar_width, label='Average Total Revenue', color='grey')
plt.bar(index + bar_width, grouped_data['Total Cost'], bar_width, label='Average Total Cost', color='orange')
plt.bar(index + 2 * bar_width, grouped_data['Total Profit'], bar_width, label='Average Total Profit', color='purple')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Average Amount')
plt.title('Average Total Revenue, Total Cost, and Total Profit by Year')
plt.xticks(index + bar_width, grouped_data.index)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

--**Insights**--

1. 2010:
Total Cost (Gray): The highest among all years, reaching close to 2 million.
Total Revenue (Orange): Also relatively high, indicating significant business activity.
Total Profit (Purple): Despite the high revenue and cost, profit is moderate, suggesting high operational costs.

2. 2011:
Significant drop in revenue, cost, and profit compared to 2010.
Costs are nearly equal to revenue, leaving very little profit.

3. 2012:
Revenue and cost have increased from 2011 levels.
Profit remains low, indicating that costs continue to rise in tandem with revenue.

4. 2013:
A noticeable drop across all metrics, possibly indicating a challenging year.
Profit is still minimal, reflecting low efficiency or high operational costs.

5. 2014-2015:
A gradual increase in revenue, cost, and profit, showing a recovery trend.
Profit margins remain low, but they are gradually improving.

6. 2016-2017:
Continued growth in revenue and costs, with 2017 showing one of the highest revenues and profits.
Although the profit is increasing, it is still relatively small compared to total revenue, highlighting high costs.
